```python
# Import the libraries for this project
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
import matplotlib.pyplot as plt 
import math 
```

```python
# Import the dataset
data = pd.read_csv("/Users/yizhang/Desktop/Jupyter Notebook data/kc_house_data.csv")
data
```

```python
data.dtypes
```

```python
# Set up our X and Y
X = data[['bedrooms','bathrooms','sqft_living','floors']]
Y = data['price']
```

```python
# Set up the training and testing
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=1/3, random_state=0)
```

```python
xtrain = np.asmatrix(xtrain)
xtest = np.asmatrix(xtest)
ytrain = np.ravel(ytrain)
ytest = np.ravel(ytest)
```

```python
# Let's see the sqft_living and the price.
plt.scatter(X['sqft_living'], Y)
plt.show()
```

```python
# Let's see the distribution of sqft_living
X['sqft_living'].hist()
plt.show()
```

```python
# Let's use xtrain and ytrain
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
pd.DataFrame(list(zip(X.columns, np.transpose(model.coef_))))
```

```python
model.intercept_
```

```python
# One house with three bedroom, two bathroom, 2500sqft, two floors into bias 1
model.predict([[3,2,2500,2]])
```

```python
# Let's check the MSE in training (Mean Square Error)
pred = model.predict(xtrain)
((pred-ytrain)*(pred-ytrain)).sum() / len(ytrain)
```

```python
# relative average deviation
(abs(pred-ytrain)/ytrain).sum() / len(ytrain)
```

```python
# Let's check the MSE in Testing
predtest = model.predict(xtest)
((predtest-ytest)*(predtest-ytest)).sum() / len(ytest)
```

```python
(abs(predtest-ytest)/ytest).sum() / len(ytest)
```
